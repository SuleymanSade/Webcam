This notebook is used for calibrating the webcam for detection. To find chessboard pattern and calibrate you can use the following code. If you want to print a chessboard pattern, check out [this website](https://markhedleyjones.com/projects/calibration-checkerboard-collection)

```python
import cv2
import json

CHECKER_BOARD:tuple = () # count in x, y
DIR = "calibration_images"
CAM_ID:str = "test"
REPLACE_CALIB:bool = True
SQUARE_SIZE = 25 #mm

with open("source/webcam.json", 'r', encoding='utf-8') as file:
    SETTINGS = json.load(file)[CAM_ID]
```

Following code detects chessboard pattern from the webcam and saves it to the DIR for calibration.

```python
import numpy as np
import os

cap = cv2.VideoCapture(SETTINGS["stream"])

if not cap.isOpened():
    raise ValueError(f'Cannot open the camera from the stream: {SETTINGS["stream"]}')

count = 0
while True:
    ret, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    ret_c, corners = cv2.findChessboardCorners(gray, CHECKER_BOARD)

    frame_c = frame.copy()
    if ret_c:
        cv2.drawChessboardCorners(frame_c, CHECKER_BOARD, corners, ret_c)
        cv2.putText(frame_c, "found pattern, saving...");

        path = os.path.join(DIR, f'image{count}.jpg')
        cv2.imwrite(path, frame)
        count+=1
    else:
        cv2.putText(frame_c, "...")
    
    cv2.imshow("Calibration", frame_c)

    key = cv2.waitKey(1) & 0xFF
    if key == ord("q"):
        break

cap.release()
```
